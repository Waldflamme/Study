
````python
def f(a, b, c, d):
    s2 = f'|{a:>10b}|{b:.>9.2f}|{c:>10.2f}|{d:<12}|'
    return s2

print(f(1, 59.06, 453, ''))
print(f(2, 159.00, 123.456, ''))
print(f(3, -15.1, -12, ''))
print(f(4, -1059.10, 1200, ''))
````
